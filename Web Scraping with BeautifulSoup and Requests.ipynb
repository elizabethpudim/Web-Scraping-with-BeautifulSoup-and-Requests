## Grabbing data from real website

We will parse **title, summary and link to video** from courses. 
https://studywebdevelopment.com/web-development-courses.html

```python
source = requests.get('https://studywebdevelopment.com/web-development-courses.html').text
```

```python
soup = BeautifulSoup(source, 'lxml')
```

```python
print(soup.prettify())
```

```python
course = soup.find('div', class_='container coursevid')
```

```python
print(course.prettify())
```

```python
coursetitle = course.p.text
```

```python
coursetitle
```

```python
summary = course.find('p', class_='palign2').text
```

```python
summary
```

```python
link = course.find('p', class_='coursecta').a
```

```python
print(link['href'])
```

```python
for course in soup.find_all('div', class_='container coursevid'):
    coursetitle = course.p.text
    print(coursetitle)
    summary = course.find('p', class_='palign2').text
    print(summary)
    link = course.find('p', class_='coursecta').a['href']
    print(link)
    print()
```

If we have some missing data (ex. no link to video or summary) let's use try/except

```python
try:
    link = course.find('p', class_='coursecta').a['href']
except Exception as e:
    link = None

print(link)
```
